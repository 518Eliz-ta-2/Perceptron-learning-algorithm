SIMPLE PERCEPTRON FOR LINEARLY SEPERABLE DATA

#classic example of supervised learning for a binary classification problem.

Dataset

```python
samples = [[1,2], [2,0], [3,1], [2,2]] #a list of 4 points, mutable
labels = [1,1,-1,-1] # +1 or -1 for classification
```

Dot product calculation Function Declaration

```python
def dot_product(a,b): #function to calculate dot product of two lists
    return sum([a[j]*b[j] for j in range(len(a))]) #returns sum of numbers inside the computed list
```

```python
w,b = [0,0] , 0 #weight (w) and bias (b) threshold
```

```python
for n in range(100): #run up to 100 iterations over the data set
    n_updates = 0 #counts how many times weights (w) and bias (b) are updated in this epoch.

    for i in range(len(labels)): #Loop over all training samples
         x,y = samples[i], labels[i]
         if y*(dot_product(w,x) + b) <= 0: #Perceptron update condition

    #Weight and bias update if misclassification occured
             w = [w[j] + y*x[j] for j in range(len(w))] #w_new = w_old + y*x
             b = b + y #b_new = b_old + y
             n_updates = n_updates + 1 #record an update occured

    if n_updates == 0: break
print("w is ",w)
print("b is ",b)
```
